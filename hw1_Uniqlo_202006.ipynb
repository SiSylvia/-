# 项目背景

优衣库为日本迅销公司的核心品牌，建立于1984年。当年是一家销售西服的小服装店，现已成为国际知名服装品牌。优衣库（Uniqlo）的内在含义是指通过摒弃不必要装潢装饰的仓储型店铺，采用超市型的自助购物方式，以合理可信的价格提供顾客希望的商品。
    
那么，如何更好地提供顾客所喜欢的商品、如何提升商品的利润、高效利用仓储空间，则是一个非常重要的企业课题。
此项目的委托为：通过分析大量的商品、用户以及销售数据，来更好地分析用户购买偏好以及不同产品的盈利情况。

# 业务问题与分析思路

业务问题的定义，采取What-Why-How的逻辑框架对业务问题进行递进式拆析，其中，What是问题的表象，Why是导致表象的深层原因，How是针对原因所提出的改善表象的对策建议。据此所定义的业务问题为：
    
What：整体商品销售情况如何？

Why：从各个维度对商品销售情况进行分析，找出各个维度上的销售特点及影响销售的原因，并据此提出业务建议。

分析思路：

1. 产品维度：
    1. 寻找对销售贡献较大的产品：销量最高、销售额最多、销售利润最大的产品分别是哪些？这些产品分别有什么特点？
    2. 寻找对销售贡献最小的产品：销量最低、销售额最少、销售利润最小的产品分别是哪些？这些产品分别有什么特点？
2. 销售渠道与销售时间维度：
    1. 线上和线下渠道分别对销售（销售、销量、利润）的贡献如何？
    2. 周末和工作日的销售（销售、销量、利润）是否有差别？
3. 消费者维度：
    1. 主要消费群体画像是怎样？不同消费者人口学变量（性别、年龄、城市）对销售的贡献如何？

How：如何针对消费者的产品偏好、购买渠道偏好以及商品特点（销量、成本、利润、仓储等）提出优化建议，进一步提升销售利润？

# 结论与业务建议

提醒：结论与业务建议主要基于本次数据样本。

## 主要结论：
1. 产品方面：
    1. 销售贡献：对总销量、总销售额、总利润贡献较大的品类为T恤、当季新品、配件；销量后三产品为裙子、毛衣、牛仔裤，销售额后三产品品类为：短裤、运动、袜子，利润后三产品品类为：运动、短裤、牛仔裤
    2. 产品特点：基于“销量”和“单价利润”两个维度分析，得到“配件”为高毛利高销量产品，“毛衣”、“裙子”为高毛利低销量品类，“当季新品”、“T恤”为中毛利高销量品类，“牛仔裤”为中毛利中销量品类，“袜子”、“短裤”为低毛利中销量品类，“运动”为低毛利低销量品类
    
    
2. 销售渠道与销售时间方面：
    1. 销售渠道：
        1. 样本数据主要来自线下门店，因此各品类产品线下渠道的销量、总销售额、总利润均比线上渠道大
        2. 作为线下门店，其线上业务的销量、销售额、利润分别占线上线下总销售的18.46%、18.36%、18.96%，原因可能为优衣库线下门店与线上电商联动较好，线下门店承担着电商销售的到店自提业务，说明线下门店的线上业务对门店销售贡献不容忽视。
    2. 销售时间：
        1. 从销量、销售额、利润三个维度分析，均有周末总量小于工作日总量的结果
        2. 但从平均值指标来看，周末的日均量均大于工作日日均量，说明优衣库门店在周末的销售效率明显比工作日高。原因可能为，比起工作日，消费者在周末会有更多的空闲时间进行门店购物。


3. 消费者方面：
   1. 性别： 
       1. 消费群体主要以女性为主，群体数量约为男性的1.78倍，对销量、销售额、利润的贡献约为男性的2.25倍、2.48倍、2.57倍
       2. 从人均水平来看，女性与男性的人均销量相等，人均销售额女性比男性高8元，人均利润女性比男性高5元
   2. 年龄：
       1. 消费主力主要为“20~39岁”群体，群体数量较大，对总销量、总销售额和总利润的贡献均较大
       2. “45~59”岁群体虽然对总销售贡献不大，但是其人均销量、人均销售额、人均利润指标显著高于其他年龄群体
   3. 城市：
       1. 消费主力主要来自深圳、武汉、杭州三个城市，群体数量较大，对总销量、总销售额和总利润的贡献均较大
       2. 从人均角度来看，城市南京对总销售贡献不大，但是其人均销量、人均销售额、人均利润指标显著高于其他城市

## 业务建议：
1. 优化产品库存，补货高销产品，促活低销产品，降低库存周期：
    1. 若仓库存量受限，可以考虑对低毛利低销量的“运动”品类、低毛利中销量的“袜子”和“短裤”品类适当减少库存
    2. 线下门店仓库要及时补货高销量产品，如配件（高毛利）、当季新品（中毛利）、T恤（中毛利）
    3. 对低销量产品多展开促销活动，如毛衣（高毛利）、裙子（高毛利）、运动（低毛利），尤其是高毛利产品
    
    
2. 加强线下门店与电商联动业务，提高周末门店促销活动频率：
    1. 通过减少配货时间、加强店铺之间商品调货能力等方式，进一步提高门店自提业务体验，以提高线下门店的线上销售
    2. 促销活动尽量选择销售效率较高的周末，同时注意周末前的补货上架工作
    
    
3. 迎合消费主力偏好，加强主力消费城市补给
    1. 产品品类补货、促销等工作与活动，可以针对主力消费群体（女性、“20~39岁”）的偏好展开
    2. 城市方面，迅销总部可以加大对深圳、武汉、杭州三个主力消费城市的资源投入

## 数据探索、清洗与预处理

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['font.sans-serif'] = 'SimHei'
```

```python
# 导入数据
df = pd.read_csv('../xiaozao2020/hw1_Uniqlo.csv')
df.head()
```

```python
# 数据有22292行，共13个属性，无空值
df.info()
```

```python
# 缺少分析所需要的利润属性，因此增加profit属性
df['profit'] = df['revenue']-df['unit_cost']*df['quant']
df.head()
```

```python
# 检查数值型变量，发现revenue和profit存在负数、unit_price存在0，为异常值，需要分别进行查看
df.describe()
# revenue为0大多是因unit_price为0，一般unit_price不为0，但也可能因活动促销(如买2送1)导致price为0，由于无法与业务员确认，无删除理由，故保留
# revenue还有一个为负值，从业务意义上判断顾客消费可能由于活动免单，但不可能倒贴，因此判定为异常，需要删除
# 查看profit异常值，发现profit为负数原因为销售价低于成本价，由于存在促销活动的可能且无法与业务员确认，因此保留不删除
```

```python
# 去掉异常值
df = df[(df['revenue'] >= 0)]

# 再次检查
df.describe()
```

思路：分析不同产品品类在总销量、总销售额、总利润、单件平均利润四个指标维度的情况，以了解不同产品品类的销售情况

## 时间维度：分析工作日和周末的销售特点
思路：分析周末和周日不同时间对销售的影响，分析指标为总销量、总销售额、总利润、日均销量、日均销售额、日均利润六个指标维度。这里假设样本数据在时间周期上均匀分布。

```python
# 构造所需数据
data = df.copy()
data = data[['wkd_ind','quant','revenue','profit']]
data = data.groupby(['wkd_ind']).sum()
#将"wkd_ind"索引字段都变为小写
data.index = data.index.str.lower()
# 分别计算周末和工作日的日均销售额、日均销量、日均利润
data.loc['per_weekday'] = data.loc['weekday']/5
data.loc['per_weekend'] = data.loc['weekend']/2
data = data.reset_index()

# 作图
plt.figure(figsize = (18,5))

# 周末和周日的总销量、日均销量比较
plt.subplot(131)
sns.barplot(x = 'wkd_ind', y = 'quant', data = data)
plt.tick_params(labelsize = 13)
plt.title('周末和周日的总销量、日均销量比较')

#周末和周日的总销售额、日均销售额比较
plt.subplot(132)
sns.barplot(x = 'wkd_ind', y = 'revenue', data = data)
plt.tick_params(labelsize = 13)
plt.title('周末和周日的总销售额、日均销售额比较')

#周末和周日的总利润、日均利润比较
plt.subplot(133)
sns.barplot(x = 'wkd_ind', y = 'profit', data = data)
plt.tick_params(labelsize = 13)
plt.title('周末和周日的总利润、日均利润比较')
```

##### 小结：
1. 从销量、销售额、利润三个维度分析，均有周末总量小于工作日总量的结果
2. 但从平均值指标来看，周末的日均量均大于工作日日均量，说明优衣库门店在周末的销售效率明显比工作日高
3. 原因分析：比起工作日，消费者在周末会有更多的空闲时间进行门店购物。
