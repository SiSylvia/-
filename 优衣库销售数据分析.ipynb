```python
plt.figure(figsize = (12,10))

# 
plt.subplot(221)
data = df.groupby(['product'])[['quant']].sum().sort_values('quant',ascending = False).reset_index()
fig = sns.barplot(x = 'product', y =  'quant', data = data, palette= 'OrRd_r')
for index, row in data.iterrows():
    fig.text(row.name, row.quant, row.quant,color='black', ha = 'center')
plt.title('fig 1.')

# 
plt.subplot(222)
data = df.groupby(['product'])[['revenue']].sum().sort_values('revenue', ascending = False).reset_index()
fig = sns.barplot(x = 'product', y = 'revenue', data = data, palette= 'OrRd_r')
for index, row in data.iterrows():
    fig.text(row.name, row.revenue, int(row.revenue), color = 'black', ha = 'center')
plt.title('fig 2.')

# 
plt.subplot(223)
data = df.groupby(['product'])[['profit']].sum().sort_values('profit', ascending = False).reset_index()
fig = sns.barplot(x = 'product', y = 'profit', data = data, palette= 'OrRd_r')
for index, row in data.iterrows():
    fig.text(row.name, row.profit, int(row.profit), color = 'black', ha = 'center')
plt.title('fig 3.')

# 
plt.subplot(224)
data = df.copy()
data['cost'] = data['unit_cost']*data['quant']
data_temp = data.groupby(['product'])[['revenue','cost','profit','quant']].sum()
data_temp['unit_profit'] = data_temp['profit']/data_temp['quant']
data_temp.sort_values(['unit_profit'], inplace = True, ascending = False)
data_temp = data_temp.reset_index()
fig = sns.barplot(x = 'product', y = 'unit_profit', data = data_temp, palette= 'OrRd_r')
for index, row in data_temp.iterrows():
    fig.text(row.name, row.unit_profit, round(row.unit_profit,2), color = 'black', ha = 'center')
plt.title('fig 4.')
```

```python
# Unkown
data = df.copy()
data = data[data['age_group'] != 'Unkown']
# '<20''>=60''19''60' 
data['age_group'] = data['age_group'].apply(lambda x: '19' if x == '<20' else x)
data['age_group'] = data['age_group'].apply(lambda x: '60' if x == '>=60' else x)

# 
data = data[['age_group','customer','quant','revenue','profit']]
data = data.groupby(['age_group']).sum()
# 
data['per_revenue'] = data['revenue']/data['customer']
data['per_quant'] = data['quant']/data['customer']
data['per_profit'] = data['profit']/data['customer']
data = data.reset_index()

# 
plt.figure(figsize = (12,15))

# 
plt.subplot(321)
fig = sns.barplot(x = 'age_group', y = 'quant', data = data, color = 'salmon')
for index, row in data.iterrows():
    fig.text(row.name, row.quant, round(row.quant, 2), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')
# 
plt.subplot(322)
fig = sns.barplot(x = 'age_group', y = 'per_quant', data = data, color =  'salmon')
for index, row in data.iterrows():
    fig.text(row.name, row.per_quant, int(row.per_quant), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')

# 
plt.subplot(323)
fig = sns.barplot(x = 'age_group', y = 'revenue', data = data, color = 'tomato')
for index, row in data.iterrows():
    fig.text(row.name, row.revenue, int(row.revenue), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')
# 
plt.subplot(324)
fig = sns.barplot(x = 'age_group', y = 'per_revenue', data = data, color = 'tomato')
for index, row in data.iterrows():
    fig.text(row.name, row.per_revenue, int(row.per_revenue), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')

# 
plt.subplot(325)
fig = sns.barplot(x = 'age_group', y = 'profit', data = data, color = 'orangered')
for index, row in data.iterrows():
    fig.text(row.name, row.profit, int(row.profit), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')
# 
plt.subplot(326)
fig = sns.barplot(x = 'age_group', y = 'per_profit', data = data, color = 'orangered')
for index, row in data.iterrows():
    fig.text(row.name, row.per_profit, int(row.per_profit), color = 'black', ha = 'center')
plt.title('')
plt.xlabel(' ')
```
